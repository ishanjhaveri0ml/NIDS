```python
import os
import random
import pandas as pd

def select_random_csv_files(folder_path, percent=10):
    # Get the list of CSV files in the folder
    csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
    
    # Calculate the number of files to select
    num_files = int(len(csv_files) * percent / 100)
    
    # Select random CSV files
    random_csv_files = random.sample(csv_files, num_files)
    
    return random_csv_files

def merge_csv_files(folder_path, csv_files, output_file):
    # Initialize an empty DataFrame to store the merged data
    merged_data = pd.DataFrame()
    
    # Iterate through each CSV file and merge into one DataFrame
    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        data = pd.read_csv(file_path)
        merged_data = pd.concat([merged_data, data], ignore_index=True)
    
    # Write the merged DataFrame to a new CSV file
    merged_data.to_csv(output_file, index=False)
    print(f"Merged CSV file saved as {output_file}")

# Example usage:
folder_path = r"D:\BRUTEFORCE_CSV\training"
output_file = r"D:\bruteforce_final\training_merge_10_percent.csv"

random_csv_files = select_random_csv_files(folder_path, percent=10)
merge_csv_files(folder_path, random_csv_files, output_file)
```

```python
import pandas as pd

def drop_columns_with_missing_values(df, threshold=0.75):
    # Calculate the percentage of missing values for each column
    missing_percent = df.isnull().mean() * 100
    
    # Get the names of columns with missing values exceeding the threshold
    columns_to_drop = missing_percent[missing_percent > threshold].index.tolist()
    
    # Drop the columns with more than threshold percentage of missing values
    df.drop(columns=columns_to_drop, inplace=True)
    
    return df

# Load your CSV file into a DataFrame
csv_file = r"D:\bruteforce_final\training_merge_10_percent.csv"
df = pd.read_csv(csv_file)

# Drop columns with more than 75% missing values
df = drop_columns_with_missing_values(df, threshold=75)

# Save the modified DataFrame to a new CSV file
output_file = r"D:\bruteforce_final\training_merge_10_percent.csv"
df.to_csv(output_file, index=False)

print("Columns with more than 75% missing values have been dropped.")
print(f"Modified CSV file saved as {output_file}")
```
